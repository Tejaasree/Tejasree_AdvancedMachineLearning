Importing the required libraries

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
import matplotlib.pyplot as plt
```

Loading and prepared Data

```python
# Load IMDB dataset with only the top 10,000 most frequent words
num_words = 10000
max_length = 500

(train_texts, train_targets), (test_texts, test_targets) = imdb.load_data(num_words=num_words)

# Pad sequences to ensure uniform input size
train_texts = sequence.pad_sequences(train_texts, maxlen=max_length)
test_texts = sequence.pad_sequences(test_texts, maxlen=max_length)
```

  
Building and Training different Models

a) Standard Model

```python
def create_standard_model():
    net = models.Sequential()
    net.add(layers.Embedding(num_words, 16, input_length=max_length))
    net.add(layers.Flatten())
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dense(1, activation='sigmoid'))

    net.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['accuracy'])
    return net

standard_model = create_standard_model()
standard_training = standard_model.fit(train_texts, train_targets, epochs=5, batch_size=512, validation_split=0.2)
```

b) single Hidden Layer Model

```python
def create_single_layer_model():
    net = models.Sequential()
    net.add(layers.Embedding(num_words, 16, input_length=max_length))
    net.add(layers.Flatten())
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dense(1, activation='sigmoid'))

    net.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['accuracy'])
    return net

single_layer_model = create_single_layer_model()
single_layer_training = single_layer_model.fit(train_texts, train_targets, epochs=5, batch_size=512, validation_split=0.2)
```

c) Three Layer Model

```python
def create_three_layer_model():
    net = models.Sequential()
    net.add(layers.Embedding(num_words, 16, input_length=max_length))
    net.add(layers.Flatten())
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dense(1, activation='sigmoid'))

    net.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['accuracy'])
    return net

three_layer_model = create_three_layer_model()
three_layer_training = three_layer_model.fit(train_texts, train_targets, epochs=5, batch_size=512, validation_split=0.2)
```

d) Model with 32 Units

```python
def create_32_units_model():
    net = models.Sequential()
    net.add(layers.Embedding(num_words, 16, input_length=max_length))
    net.add(layers.Flatten())
    net.add(layers.Dense(32, activation='relu'))
    net.add(layers.Dense(32, activation='relu'))
    net.add(layers.Dense(1, activation='sigmoid'))

    net.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['accuracy'])
    return net

model_32_units = create_32_units_model()
training_32_units = model_32_units.fit(train_texts, train_targets, epochs=5, batch_size=512, validation_split=0.2)
```

e)Model with 64 Units

```python
def create_64_units_model():
    net = models.Sequential()
    net.add(layers.Embedding(num_words, 16, input_length=max_length))
    net.add(layers.Flatten())
    net.add(layers.Dense(64, activation='relu'))
    net.add(layers.Dense(64, activation='relu'))
    net.add(layers.Dense(1, activation='sigmoid'))

    net.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['accuracy'])
    return net

model_64_units = create_64_units_model()
training_64_units = model_64_units.fit(train_texts, train_targets, epochs=5, batch_size=512, validation_split=0.2)
```

f) Model with MSE Loss

```python
def create_mse_loss_model():
    net = models.Sequential()
    net.add(layers.Embedding(num_words, 16, input_length=max_length))
    net.add(layers.Flatten())
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dense(1, activation='sigmoid'))

    net.compile(optimizer='rmsprop',
                loss='mse',
                metrics=['accuracy'])
    return net

mse_loss_model = create_mse_loss_model()
mse_loss_training = mse_loss_model.fit(train_texts, train_targets, epochs=5, batch_size=512, validation_split=0.2)
```

g) Model with Tanh Activation

```python
def create_tanh_model():
    net = models.Sequential()
    net.add(layers.Embedding(num_words, 16, input_length=max_length))
    net.add(layers.Flatten())
    net.add(layers.Dense(16, activation='tanh'))
    net.add(layers.Dense(16, activation='tanh'))
    net.add(layers.Dense(1, activation='sigmoid'))

    net.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['accuracy'])
    return net

tanh_model = create_tanh_model()
tanh_training = tanh_model.fit(train_texts, train_targets, epochs=5, batch_size=512, validation_split=0.2)
```

h) Model with dropout

```python
def create_dropout_model():
    net = models.Sequential()
    net.add(layers.Embedding(num_words, 16, input_length=max_length))
    net.add(layers.Flatten())
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dropout(0.5))
    net.add(layers.Dense(16, activation='relu'))
    net.add(layers.Dropout(0.5))
    net.add(layers.Dense(1, activation='sigmoid'))

    net.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['accuracy'])
    return net

dropout_model = create_dropout_model()
dropout_training = dropout_model.fit(train_texts, train_targets, epochs=5, batch_size=512, validation_split=0.2)
```

  Plotting the accuracy trends

```python
def visualize_results(histories, key='val_accuracy'):
    plt.figure(figsize=(12, 8))

    for label, history in histories:
        plt.plot(range(1, len(history.history[key]) + 1), history.history[key], label=label)

    plt.xlabel('Epochs')
    plt.ylabel(key)
    plt.legend()
    plt.show()

histories = [
    ('Standard', standard_training),
    ('Single Layer', single_layer_training),
    ('Three Layers', three_layer_training),
    ('32 Units', training_32_units),
    ('64 Units', training_64_units),
    ('MSE Loss', mse_loss_training),
    ('Tanh Activation', tanh_training),
    ('Dropout', dropout_training)
]

visualize_results(histories)
```

Evaluating the models of test Data

```python
def assess_model(network, x_test, y_test):
    test_loss, test_accuracy = network.evaluate(x_test, y_test)
    return test_loss, test_accuracy

results = {
    "Standard": assess_model(standard_model, test_texts, test_targets),
    "Single Layer": assess_model(single_layer_model, test_texts, test_targets),
    "Three Layers": assess_model(three_layer_model, test_texts, test_targets),
    "Tanh Activation": assess_model(tanh_model, test_texts, test_targets),
    "Dropout": assess_model(dropout_model, test_texts, test_targets),
}

for name, (loss, acc) in results.items():
    print(f"{name} Model - Loss: {loss:.4f}, Accuracy: {acc:.4f}")
```
